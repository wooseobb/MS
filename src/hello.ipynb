```python
# 두 수를 입력받아 곱 계산
a = int(input("첫 번째 정수 입력: "))
b = int(input("두 번째 정수 입력: "))

print('{0} * {1} = {2:,}'.format(a, b, a*b))
```

```python
height_cm = float(input("키 입력(cm): "))
weight = float(input("몸무게 입력(kg): "))

height_m = height_cm / 100
bmi = weight / (height_m ** 2)

if bmi < 18.5:
    status = "저체중"
elif bmi < 23:
    status = "정상체중"
elif bmi < 25:
    status = "과체중"
else:
    status = "비만"

print(f"BMI 지수는 {bmi:.2f}로 '{status}'입니다.")
```

```python
todo_list = []

while True:
    print("\n메뉴 선택(1:추가, 2:삭제, 0:종료)")
    menu = input(">> ")

    if menu == '1':
        task = input("할 일 입력 >> ")
        todo_list.append(task)

    elif menu == '2':
        if not todo_list:
            print("삭제할 항목이 없습니다.")
            continue

        for i, task in enumerate(todo_list, 1):
            print(f"{i}. {task}")

        idx = int(input("삭제할 번호 입력 >> ")) - 1

        if 0 <= idx < len(todo_list):
            todo_list.pop(idx)
        else:
            print("잘못된 번호입니다.")

    elif menu == '0':
        print("프로그램 종료")
        break

    else:
        print("잘못된 입력입니다.")

    print("\n현재 할 일 목록")
    for i, task in enumerate(todo_list, 1):
        print(f"{i}. {task}")
```

```python
phone_book = {}

while True:
    print("\n메뉴 선택(1:추가, 2:검색, 0:종료)")
    menu = input(">> ")

    if menu == '1':
        name = input("이름 입력 >> ")
        number = input("전화번호 입력 >> ")
        phone_book[name] = number

    elif menu == '2':
        name = input("검색할 이름 입력 >> ")

        if name in phone_book:
            print(f"{name} : {phone_book[name]}")
        else:
            print("등록되지 않은 이름입니다.")

    elif menu == '0':
        print("프로그램 종료")
        break

    else:
        print("잘못된 입력입니다.")
```
