```python
# 두 수를 입력받아 곱 계산
a = int(input("첫 번째 정수 입력: "))
b = int(input("두 번째 정수 입력: "))

print('{0} * {1} = {2:,}'.format(a, b, a*b))
```

```python
# 두 실수를 입력받아 곱 계산
a = float(input("첫 번째 값 입력: "))
b = float(input("두 번째 값 입력: "))

print(f'{a:.2f} * {b:.3f} = {a*b:,.2f}')
```

#도전 문제: 원의 둘레 & 넓이

```python
r = float(input("반지름 입력(cm): "))

print(f"반지름 : {r:.1f}cm")
print(f"원의 둘레 : {2*3.14*r:.1f}cm")
print(f"원의 넓이 : {3.14*r*r:,.1f}㎠")
```

```python
score = int(input("점수 입력: "))

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
birth_last = int(input("출생연도 끝자리 입력: "))
age = int(input("나이 입력: "))

if age >= 65:
    print("언제든 구매 가능합니다.")
elif birth_last in [1, 6]:
    print("월요일")
elif birth_last in [2, 7]:
    print("화요일")
elif birth_last in [3, 8]:
    print("수요일")
elif birth_last in [4, 9]:
    print("목요일")
elif birth_last in [5, 0]:
    print("금요일")
else:
    print("잘못된 입력입니다.")
```

```python
day = int(input("오늘은 몇 일?(1~31): "))
car = input("차량번호 4자리 입력: ")

if day < 1 or day > 31:
    print("일(day) 입력 오류입니다.")
elif len(car) != 4 or not car.isdigit():
    print("차량번호 입력 오류입니다.")
else:
    last_digit = int(car[-1])

    if day % 2 == last_digit % 2:
        print("운행 가능합니다.")
    else:
        print("운행 불가능합니다.")
```

```python
height_cm = float(input("키 입력(cm): "))
weight = float(input("몸무게 입력(kg): "))

height_m = height_cm / 100
bmi = weight / (height_m ** 2)

if bmi < 18.5:
    status = "저체중"
elif bmi < 23:
    status = "정상체중"
elif bmi < 25:
    status = "과체중"
else:
    status = "비만"

print(f"BMI 지수는 {bmi:.2f}로 '{status}'입니다.")
```
